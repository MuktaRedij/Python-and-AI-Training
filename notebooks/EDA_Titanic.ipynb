This notebook performs EDA (Exploratory Data Analysis) on the Titanic dataset using Python.  
We analyze survival patterns based on passenger gender, age, and travel class.  
The goal is to identify insights and trends from real-world historical data.


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load Titanic dataset built into seaborn
df = sns.load_dataset("titanic")

df.head()
```

```python
df.info()
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['survived'].value_counts().plot(kind='bar', color=['red','green'])
plt.title("Survival Count")
plt.xlabel("0 = Did Not Survive, 1 = Survived")
plt.ylabel("Count")
plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x='class', hue='survived', data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

## ðŸ“Ž Key Insights

- More passengers **did not survive** compared to those who survived.
- **Females had a significantly higher survival rate** than males.
- **First-class passengers survived more often** than second and third class passengers.
- Several columns contain missing values, especially `age`.

This indicates survival was influenced by **gender**, **social class**, and possibly age.
